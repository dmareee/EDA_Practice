### **2. Memeriksa value jauh dari dataset (Outlier)**

````python
df = df.drop(columns=['ID'])
````

````python
df.columns
````

````python
import matplotlib.pyplot as plt

cols = ['Cost_of_the_Product', 'Prior_purchases', 'Product_importance',
        'Discount_offered']
df[cols].boxplot(figsize=(30,15), fontsize=15)
plt.title("Check outlier" , fontsize=30)
plt.show()
````

````python
# Cek Outlier dengan IQR Outlier
def outlier_iqr(data):
    outliers = []
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    for i in data:
        if i < lower_bound or i > upper_bound:
            outliers.append(i)
    return outliers

numeric_cols = df[cols].select_dtypes(include='number')

data_outlier = {}
for col in numeric_cols:
    data_outlier[col] = outlier_iqr(df[col])
    print('Outlier (' + col + '):', len(data_outlier[col]), 'outliers')
````

````python
def handling_outlier(x):
    q1 = x.quantile(0.25)
    q3 = x.quantile(0.75)
    iqr = q3 - q1
    lower_limit = q1 - 1.5 * iqr
    higher_limit = q3 + 1.5 * iqr
    print('lower_limit:', lower_limit)
    print('higher_limit:', higher_limit)
    x = np.where(x > higher_limit, higher_limit, x)
    x = np.where(x < lower_limit, lower_limit, x)
    return x
````

````python
df.Discount_offered = handling_outlier(df.Discount_offered)
df.Prior_purchases = handling_outlier(df.Prior_purchases)

cols = ['Cost_of_the_Product', 'Prior_purchases', 'Product_importance',
        'Discount_offered']
df[cols].boxplot(figsize=(30,15), fontsize=15)
plt.title("Check outlier" , fontsize=30)
plt.show()
````

### **4. Hyperparameter Tuning**

- Hyperparameter Tuning adalah proses mencari kombinasi terbaik dari hyperparameter untuk model pembelajaran mesin.
- Hyperparameter adalah parameter yang didefinisikan sebelum training model dan tidak ditentukan oleh model itu sendiri.
- Proses tuning ini dilakukan untuk meningkatkan kinerja model dengan mencari kombinasi hyperparameter yang optimal.
- Metode tuning yang umum digunakan termasuk Grid Search, Random Search, dan Bayesian Optimization.

Dengan memahami karakteristik dan penggunaan dari masing-masing algoritma serta teknik tuning hyperparameter, dapat memilih dan mengoptimalkan model sesuai dengan kebutuhan dan dataset yang dimiliki.

## Teori


### Penjelasan Singkat
Plotly library pada Python adalah sebuah alat *open-source* yang digunakan untuk visualisasi data serta memahami data dengan mudah dan sederhana. Plotly mendukung berbagai macam bentuk grafik, seperti line chart, scatter plot, histogram, box plot, dan lain-lain. Beberapa kelebihan library plotly antara lain:

- **Hover Tool**: Plotly mempunyai fitur hover tool yang dapat membantu kita mendeteksi outlier atau anomali pada data yang besar.
- **Visual Appeal**: Plotly mempunyai visual yang menarik dan dapat diterima oleh semua kalangan usia.
- **Endless Customization**: Secara umum, Plotly mengizinkan kita melakukan kostumisasi secara meluas terhadap grafik yang ingin kita visualisasikan agar lebih mudah dipahami oleh orang lain.

### Instalasi

Sebelum dapat menggunakan library Plotly, kita perlu melakukan instalasi Plotly dengan cara:

```
pip install plotly
```

lalu import library tersebut agar dapat kita gunakan dengan kode berikut:


````python
import plotly.graph_objects as go
import plotly.express as px
````

Note:
- go digunakan secara umum sebagai alias untuk modul `plotly.graph_objects`, modul ini adalah implementasi plotly dengan low interface yang **lebih rumit daripada `plotly.express` namun lebih fleksibel.**
- px digunakan secara umum sebagai alias untuk modul `plotly.express`, modul ini adalah implementasi dari plotly dengan high interface yang **lebih mudah sederhana dan mudah digunakan ketimbang `plotly.graph_objects`**
- Dalam materi kali ini, akan lebih banyak digunakan plotly express dalam implementasi plotly.

### Perbandingan Syntax plotly.graph_objects dan plotly.express

````python
# Data
x = [1, 2, 3, 4]
y = [10, 11, 12, 13]

# Membuat grafik garis
fig_go = go.Figure(data=go.Scatter(x=x, y=y, mode='lines')) # Menggunakan plotly.graph_objects
fig_px = px.line(x=x, y=y)                                  # Menggunakan plotly.express

# Menampilkan grafik
fig_go.show(renderer='colab')
fig_px.show(renderer='colab')
````

Terlihat bahwa dengan syntax yang lebih pendek dalam pembuatan grafik, `plotly express` dapat membuat grafik yang serupa dengan grafik yang menggunakan `plotly.graph_objects`.

### Renderer
Pada implementasi plotly, ketika kita ingin melakukan render atau menampilkan grafik dengan fungsi `.show()`, sebaiknya ditetapkan parameter `renderer="<opsi-render>"` untuk menentukan bagaimana grafik akan dirender.

Berikut adalah beberapa opsi render yang bisa digunakan:
- "notebook": Menampilkan plot di dalam notebook Jupyter jika kita sedang menggunakan Jupyter Notebook atau JupyterLab.
- "browser": Menampilkan plot dalam jendela browser yang terpisah dari notebook atau kode Python.
- "iframe": Menampilkan plot dalam bentuk iframe yang dapat disematkan di situs web atau dokumentasi online.
- "colab": Menggunakan renderer ini saat kita bekerja di Google Colab.
- "png", "jpeg", "webp": Menghasilkan gambar PNG, JPEG, atau WebP dari plot tanpa menampilkan secara interaktif.

Contoh penggunaan: `fig.show(renderer="notebook")`

## Contoh Penggunaan Dasar

### 1. Line Chart menggunakan Plotly

````python
# Definisikan fitur atau baris dan kolom untuk digunakan pada grafik
x = [1,2,3,4,5]
y = [2,8,6,4,10]

# Definisikan jenis chart dan sertakan parameter seperti baris (x), kolom (y), dan judul grafik
fig = px.line( x = x ,
              y = y,
              title = 'Contoh Line Chart Sederhana menggunakan Plotly')

# Sesuaikan ukuran chart
fig.update_layout(width=800, height=600)

# Tampilkan chart yang sudah didefinisikan
fig.show(renderer="colab")
````

````python
## Contoh jika ditambah parameter lain untuk kustomisasi yang memungkinkan di Plotly

fig = px.line(x=x, y=y, title='Kustomisasi Line Chart menggunakan Plotly',
              color_discrete_sequence=['blue'], labels={'x': 'Sumbu X', 'y': 'Sumbu Y'},
              line_shape='spline', template='plotly_dark')

fig.update_layout(width=800, height=600)
fig.show(renderer="colab")
````

### 2. Bar Chart menggunakan Plotly

````python
# Data sederhana
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 78, 34]

# Definisikan bar chart
fig_simple = px.bar(x=categories, y=values, title='Bar Chart Sederhana')

# Tampilkan bar chart
fig_simple.show(renderer="colab")
````

````python
# Bar chart dengan kustomisasi
fig_customized = px.bar(x=categories, y=values, title='Bar Chart dengan Kustomisasi',
                        color=values, color_continuous_scale='Bluered',
                        labels={'x': 'Kategori', 'y': 'Nilai'}, template='plotly_dark')

fig_customized.show(renderer="colab")
````

### 3. Histogram menggunakan Plotly

````python
# Import numpy untuk mempermudah pembuatan data
import numpy as np

# Data random menggunakan numpy
x = np.random.randn(100) # Membuat array berukuran 100 terdiri dari angka random dengan distribusi normal

# Definisikan histogram
fig_simple = px.histogram(x=x, title='Histogram Sederhana')

# Tampilkan histogram
fig_simple.show(renderer="colab")
````

````python
# Histogram dengan kustomisasi
fig_customized = px.histogram(x=x, title='Histogram dengan Kustomisasi',
                              color_discrete_sequence=['black'],
                              opacity=0.8,
                              labels={'x': 'Nilai'},
                              template='plotly_white',)

fig_customized.show(renderer="colab")
````

### 4. Box Plot menggunakan Plotly

````python
# Data sederhana
np.random.seed(0)
x = np.random.randn(100)

# Definisikan box plot
fig_simple = px.box(x=x, title='Box Plot Sederhana')

# Tampilkan box plot
fig_simple.show(renderer="colab")
````

````python
# Box plot dengan kustomisasi
fig_customized = px.box(x=x, title='Box Plot dengan Kustomisasi',
                        color_discrete_sequence=['black'],
                        template='seaborn',
                        points='all')
fig_customized.show(renderer="colab")
````

### 5. Pie Chart menggunakan Plotly

````python
# Data sederhana
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [15, 30, 25, 10, 20]

# Definisikan pie chart
fig_simple = px.pie(labels=labels, values=sizes, title='Pie Chart Sederhana')

# Tampilkan pie chart
fig_simple.show(renderer="colab")
````

````python
# Pie chart dengan kustomisasi
fig_customized = px.pie(labels=labels, values=sizes, title='Pie Chart dengan Kustomisasi',
                        color=labels, color_discrete_map={'A': 'skyblue', 'B': 'salmon', 'C': 'lightgreen', 'D': 'gold', 'E': 'lightcoral'},
                        hole=0.3)

fig_customized.show(renderer="colab")
````
